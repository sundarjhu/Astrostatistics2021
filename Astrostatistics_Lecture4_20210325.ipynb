##Let us compute the mean of $N$ Standard Uniform random numbers, gradually increasing $N$.
##Each time, we will compute the mean a large number of times so we sample the **sampling distribution** at high-enough resolution.
##We will then compare the result with the PDF of a Normal distribution that has the same mean and the expected standard deviation.
###Recall: mean of a Standard Uniform random number: 0.5. Standard deviation: $\displaystyle{1\over \sqrt{12}}$.
###Corresponding properties of the sample mean of N Standard Uniform random numbers: mean = 0.5, standard deviation = $\displaystyle{1\over \sqrt{12N}}$.

```python
import numpy as np
from scipy.stats import uniform, norm
import matplotlib.pyplot as plt

N = [2, 3, 10] #numbers to sum
samples = 10000 #number of times each sum is sampled

fig, ax = plt.subplots(ncols = 3, nrows = 1, constrained_layout = True, clear = True, figsize = (12, 4))
axf = ax.ravel()
for i in range(len(N)):
  x = uniform.rvs(size = (N[i], samples))
  xsum = x.mean(axis = 0)
  axf[i].hist(xsum, bins = 50, histtype = 'step', density = True)
  lims = axf[i].set_xlim()
  xx = np.linspace(lims[0], lims[1], 100)
  axf[i].plot(xx, norm.pdf(xx, loc = 0.5, scale = 1 / np.sqrt(12 * N[i])))
  axf[i].set_title('N = {}'.format(N[i]))
```

#Even for $N\approx 5$, the Normal approximation for the sample mean is pretty good!

##Standardisation: how to standardise if the population mean is unknown
###$Z = \displaystyle{X-\overline{X}\over\sigma}\Longrightarrow Z\sim\mathscr{N}\Bigg(0, 1 + \displaystyle{1 \over N}\Bigg)$

```python
#Verify the above claim by drawing N samples of X from a Gaussian, computing the sample mean, then computing the Z score
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

N = 20 #number of samples
mu = 15 #we will pretend the population mean is unknown
sigma = 3
x = norm.rvs(loc = mu, scale = sigma, size = N)

xbar = x.mean()
z = (x - xbar) / sigma
plt.figure(figsize = (8, 8))
_ = plt.hist(z, histtype = 'step', density = True)
zz = np.linspace(min(z), max(z), 1000)
_ = plt.plot(zz, norm.pdf(zz), 'r--', alpha = 0.2, label = 'sigma = 1')
_ = plt.plot(zz, norm.pdf(zz, scale = 1 + 1 / N), label = 'sigma = 1 + 1/N')

plt.legend(loc = 'best')
```
